```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
V = np.linspace(-8, 8, 16001) #[V]
```

### Potasio (K+)

```python
K_out = 20 #[mM] 
K_in = 430 #[mM]

u = np.exp(V)

jK = V*(K_out - K_in*u)/(1-u)

plt.plot(V, jK, label = "Potasio (K+)")
plt.axhline(0, ls = '--', lw = '0.5', c = 'k')
plt.axvline(0, ls = '--', lw = '0.5', c = 'k')
plt.legend()
plt.show()
```

### Sodio (Na+)

```python
Na_out = 440 #[mM] 
Na_in = 50 #[mM]

u = np.exp(V)

jNa = V*(Na_out - Na_in*u)/(1-u)

plt.plot(V, jNa, label = "Sodio (Na+)")
plt.axhline(0, ls = '--', lw = '0.5', c = 'k')
plt.axvline(0, ls = '--', lw = '0.5', c = 'k')
plt.legend()
plt.show()
```

### Cloro (Cl-)

```python
Cl_out = 550 #[mM] 
Cl_in = 65 #[mM]

u = np.exp(-V)

jCl = V*(Cl_out - Cl_in*u)/(1-u)

plt.plot(V, jCl, label = "Cloro (Cl-)")
plt.axhline(0, ls = '--', lw = '0.5', c = 'k')
plt.axvline(0, ls = '--', lw = '0.5', c = 'k')
plt.legend()
plt.show()
```

### Calcio (Ca2+)

```python
Ca_out = 4 #[mM] 
Ca_in = 0.1/1000 #[mM]

u = np.exp(2*V)

jCa = V*(Ca_out - Ca_in*u)/(1-u)

plt.plot(V, jCa, label = "Calcio (Ca2+)")
plt.axhline(0, ls = '--', lw = '0.5', c = 'k')
plt.axvline(0, ls = '--', lw = '0.5', c = 'k')
plt.legend()
plt.show()
```

#### Grafico todos juntos

```python
plt.plot(V, jK, label = "Potasio (K+)")
plt.plot(V, jNa, label = "Sodio (Na+)")
plt.plot(V, jCl, label = "Cloro (Cl-)")
plt.plot(V, jCa, label = "Calcio (Ca2+)")
plt.axhline(0, ls = '--', lw = '0.5', c = 'k')
plt.axvline(0, ls = '--', lw = '0.5', c = 'k')
plt.legend()
plt.show()
```
